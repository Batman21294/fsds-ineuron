```python
df_titanic = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
```

```python
df_titanic
```

```python
def check_Pclass(a):
    if a<3:
        return "poor"
    else:
        return "rich"
```

```python
check_Pclass(1)
```

```python
check_Pclass(3)
```

```python
df_titanic['poor or rich'] = df_titanic['Pclass'].apply(check_Pclass)
```

```python
df_titanic
```

```python
df_titanic['length of passenger name'] = df_titanic['Name'].apply(lambda x:len(x))
```

```python
df_titanic
```

```python
df_titanic.drop('length of passenger name',axis=1,inplace=True)
```

```python
df_titanic
```

```python
df_titanic['length of name'] = df_titanic['Name'].apply(len)
```

```python
df_titanic
```

```python
df_titanic["square of fare"] = df_titanic['Fare'].apply(lambda x:x**2)
```
